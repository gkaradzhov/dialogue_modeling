```python
import pandas as pd
```

```python
dataset = pd.read_csv('sample_data_output - line_annotation_reversed.tsv', delimiter='\t')
```

```python
dataset = dataset.fillna('0')
```

```python
dataset = dataset.rename({'Unnamed: 2': 'bot'}, axis=1)
```

```python
dataset
```

```python
processed_dialogues = []

current_dialogue = []
for item in dataset.iterrows():
    if item[1]['origin'] == '0' and len(current_dialogue) > 0:
        processed_dialogues.append(current_dialogue)
        current_dialogue = []
    current_dialogue.append({'content': item[1]['content'], 'bot': item[1]['bot'],
                             'labels': [b.strip() for b in item[1]['Type'].split(',')]})
    
```
